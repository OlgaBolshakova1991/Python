# Выбор данных

```python
# Выбор 1 столбца - [выжил или нет]
df['Survived']
```

```python
# Выбор нескольких столбцов [выжил или нет, в каком классе путешествовал человек]
df[['Survived', 'Pclass']]
```

```python
# Выбор определенного кол-ва рядов
# Синтаксис df[df[col] !=|==|>|<| значение]
df[df['Age'] < 20]
```

```python
# Для отбора можно использовать несколько условий одновременно
# Знак & означает 'and', а знак | 'or'
# Выживших детей было 65 человек до 18 лет и 65 детей погибло.
df[(df['Age'] < 18) & (df['Survived'] > 0)] 
```

```python
df[(df['Age'] < 18) & (df['Survived'] < 1)] 
```

```python
# Выбор определенного кол-ва рядов и столбцов
# используется метод loc в [], первый аргумент индекс или селектор, а второй - список со столбцами
# 545 человек погибло из 887
df.loc[df['Survived'] < 1, ['Age']]
```

# Простая статистика

Pandas позволяет получить основные простые данные для описательной статистики

Такие как минимальное значение в столбце, максимальное значение, сумма всех значений, среднее значение

```python
# Максимальное значение
print(df['Age'].max()) # Максимальный возраст человека на борту 80.0 лет
# Минимальное значение
print(df['Age'].min()) # Минимальный возраст человека на борту 4 месяца 
# Среднее значение
print(df['Age'].mean()) # Средний возраст человека на борту 29 лет
# Сумма
print(df['Fare'].sum()) # Было потрачено денег проезд  28655 фунтa
```

Эту же статистику можно рассчитывать сразу для нескольких столбцов

```python
# Медианное значение
df[['Fare', 'Age']].median()
```
